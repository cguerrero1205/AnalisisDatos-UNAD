```python
# Cargamos a nuestro entorno de trabajo las librerias necesarias.

import matplotlib.pyplot as plt
from sklearn import linear_model # usando sklear para saber los valores optimos
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
# Cargamos la información del CSV en una variable llamada data

data = pd.read_csv("C:/Users/USUARIO/OneDrive/UNAD/Ing de Sistemas/3er semestre (2023-16-04)/ANÁLISIS DE DATOS/Etapa 2/regresionLineal.csv", sep=",")
```

```python
#Realizo la grafica de dispersión
data.plot.scatter(x="metro", y="precio")
plt.show()
```

```python
# Ejemplo Regresión Lineal
 
regresion = linear_model.LinearRegression()
```

```python
#Agrego los datos en un array o vector
years = data["metro"].values.reshape((-1,1))
```

```python
#Ahora si creamos el modelo
modelo = regresion.fit(years, data["precio"])
```

```python
print("Interseccion (b)", modelo.intercept_)
#imprimos la pendiente
print("Pendiente (m)", modelo.coef_)
```

```python
#Para comprobar el modelo y su error
entrada= [[5],[15],[20],[25]]
predicciones = modelo.predict(entrada)
print(predicciones)
```

```python
#Se compara los puntos de la predicción con los valores reales
data.plot.scatter(x="metro", y="precio", label='Datos originales')
plt.scatter(entrada, predicciones, color='red')
plt.plot(entrada, predicciones, color='black', label='Línea de regresión')
plt.xlabel('metro')
plt.ylabel('precio')
plt.legend()
plt.show()
```

```python
#Se compara los valores de las predicciones con los reales
data['prediccion'] = predicciones
data['error'] = predicciones - data['precio']


data.head()
```
