```python
#Generar grafica de puntos
data[['BMI','diabetes']].plot.scatter(x='BMI',y='diabetes')
```

```python
# Vamos a pintar una función logistica sobre estos datos

# pruebas de parametro
w = 0.09
b = -3.6
```

```python
# puntos de la recta
x = np.linspace(0,data['BMI'].max(),100)
y = 1/(1+np.exp(-(w*x+b)))
```

```python
# grafica de la recta
data.plot.scatter(x='BMI',y='diabetes')
plt.plot(x, y, color='black')
plt.ylim(0,data['diabetes'].max()*1.1)
plt.scatter(x, y, color='#A9E2F3')
# plt.grid()
plt.xlabel('Indice de Masa Corporal')
plt.ylabel('Diabetes 1:Positivo 0:Negativo')
plt.show()
```
